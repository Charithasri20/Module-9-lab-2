# Data preparation

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Load the iris dataset
iris = load_iris()
X = iris['data']
y = iris['target']
names = iris['target_names']
feature_names = iris['feature_names']

# Scale data to have mean 0 and variance 1
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data set into training and testing
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=2)
```

# Visualising dataset

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
for target, target_name in enumerate(names):
    X_plot = X[y == target]
    ax1.plot(X_plot[:, 0], X_plot[:, 1],
             linestyle='none',
             marker='o',
             label=target_name)
ax1.set_xlabel(feature_names[0])
ax1.set_ylabel(feature_names[1])
ax1.axis('equal')
ax1.legend();

for target, target_name in enumerate(names):
    X_plot = X[y == target]
    ax2.plot(X_plot[:, 2], X_plot[:, 3],
             linestyle='none',
             marker='o',
             label=target_name)
ax2.set_xlabel(feature_names[2])
ax2.set_ylabel(feature_names[3])
ax2.axis('equal')
ax2.legend();
```

# MLP for multiclass classification

```python
import torch
import torch.nn.functional as F
import torch.nn as nn
from torch.autograd import Variable
import tqdm
```

```python
# Defining the model architecture
class Model(nn.Module):
    def __init__(self, input_dim):
        super(Model, self).__init__()
        self.layer1 = nn.Linear(input_dim, 50)
        self.layer2 = nn.Linear(50, 50)
        self.layer3 = nn.Linear(50, 3)

    def forward(self, x):
        x = F.relu(self.layer1(x))
        x = F.relu(self.layer2(x))
        x = F.softmax(self.layer3(x), dim=1)
        return x
```

```python
# Instantiating the model, using Adam optimiser, and Cross Entropy Loss, which is quite commonlu used for classification tasks.
model     = Model(X_train.shape[1])
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
loss_fn   = nn.CrossEntropyLoss()
```

# Train the model

```python
# Train for 100 epochs
EPOCHS  = 100
X_train = torch.from_numpy(X_train).float()
X_test= torch.from_numpy(X_test).float()
y_test = torch.from_numpy(y_test)
y_train = torch.from_numpy(y_train)


loss_list     = np.zeros((EPOCHS,))
accuracy_list = np.zeros((EPOCHS,))

for epoch in tqdm.trange(EPOCHS):
    y_pred = model(X_train)
    loss = loss_fn(y_pred, y_train)
    loss_list[epoch] = loss.item()

    # Zero gradients
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    with torch.no_grad():
        y_pred = model(X_test)
        correct = (torch.argmax(y_pred, dim=1) == y_test).type(torch.FloatTensor)
        accuracy_list[epoch] = correct.mean()
```

# Plot training progress

```python
fig, (ax1, ax2) = plt.subplots(2, figsize=(12, 6), sharex=True)

ax1.plot(accuracy_list)
ax1.set_ylabel("validation accuracy")
ax2.plot(loss_list)
ax2.set_ylabel("validation loss")
ax2.set_xlabel("epochs");
```

Number of Hidden Layers:
Increasing the number of hidden layers can sometimes improve the network's ability to learn complex patterns, especially in deep learning tasks. However, adding too many layers can lead to overfitting if not properly regularized.
Conversely, reducing the number of hidden layers may simplify the network and reduce training time, but it may also limit its ability to learn complex relationships in the data.
Number of Neurons in the Hidden Layer:
Increasing the number of neurons in a hidden layer can increase the model's capacity to learn complex patterns in the data. However, too many neurons can lead to overfitting, especially if the dataset is small.
Conversely, decreasing the number of neurons can help prevent overfitting and reduce computational complexity, but it may also limit the model's ability to capture intricate patterns.
Activation Function:
The choice of activation function can greatly influence the model's ability to learn and generalize. Common activation functions include ReLU (Rectified Linear Unit), sigmoid, tanh (Hyperbolic Tangent), and softmax (for multi-class classification).
ReLU is often a good default choice for hidden layers due to its simplicity and effectiveness in combating the vanishing gradient problem.
Sigmoid and tanh functions are commonly used in the output layer for binary classification tasks.
Softmax is typically used in the output layer for multi-class classification tasks.
When making these adjustments, it's essential to monitor the model's performance on a validation set to ensure that changes improve generalization and don't lead to overfitting. Additionally, techniques such as dropout regularization and early stopping can be employed to mitigate overfitting and fine-tune the model's architecture.
